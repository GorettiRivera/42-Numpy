### 5. Using the function in Q3, get m vectors and bind them together (to have a m x n) matrix. Plot the matrix with matplotlibâ€™s imshow

```python
m = 500
n = 700
arr = np.full((m, n), dim(n))
matp = np.matrix(arr)
matp
```

```python
plt.imshow(matp)
```

### 6. Multiply the resulting matrix from Q5 with the matrix of a picture of your choice. Plot the resulting matrix.

```python
img1 = img.imread("land.png")
plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.imshow(img1, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.title('Original')

for i in range(0, 3):
    img1[:,:,i] *= matp

plt.subplot(2,2,2)
plt.imshow(img1)
plt.axis('off')
plt.title('Image transformed')
```

# <font color="blue"> Part II </font>

```python
image_name='land.png'

im_data = img.imread(image_name)

# Data in MxNx3 (RGB)
red_data = im_data[:,:,0]
green_data = im_data[:,:,1]
blue_data = im_data[:,:,2]

# swap RGB order
image1 = np.array([blue_data, green_data, red_data])
image1 = np.swapaxes(image1, 0, 1)
image1 = np.swapaxes(image1, 1, 2)

# colour swapped
limg3=list(im_data)
swapped=[]
for i in limg3:
  temp=[]
  for j in i:
    temptemp=[]
    temptemp.append(j[1])
    temptemp.append(j[2])
    temptemp.append(j[0])
    temp.append(temptemp)
  swapped.append(temp)
  

cmap = 'gray'

plt.figure(figsize=(20,20))

plt.subplot(4,2,1)
plt.imshow(im_data, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('Original')

plt.subplot(4,2,2)
plt.imshow(image1, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('BGR')

plt.subplot(4,2,3)
plt.imshow(swapped, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('Swapped')

plt.subplot(4,2,4)
plt.imshow(red_data, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('Red')

plt.subplot(4,2,5)
plt.imshow(green_data, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('Green')

plt.subplot(4,2,6)
plt.imshow(blue_data, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.title('Blue')
```
